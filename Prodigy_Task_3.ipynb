```python
file_name = "story.txt"

try:
    with open(file_name, 'r', encoding='utf-8') as file:
        corpus = file.read()
except FileNotFoundError:
    print(f"Error: The file '{file_name}' was not found. Please make sure you have uploaded it.")
    exit()
```

```python
# Split the corpus into a list of words
words = corpus.lower().split()

# Build the Markov chain dictionary
markov_chain = {}

for i in range(len(words) - 1):
    current_word = words[i]
    next_word = words[i+1]

    # If the current word isn't in the dictionary, add it with an empty list
    if current_word not in markov_chain:
        markov_chain[current_word] = []

    # Append the next word to the list
    markov_chain[current_word].append(next_word)

print(markov_chain)
```

```python
import random

# Choose a random starting word
current_word = random.choice(list(markov_chain.keys()))
generated_text = [current_word]

# Define how many words you want to generate
num_words_to_generate = 20

# Generate the text
for _ in range(num_words_to_generate):
    # If the current word is not in the dictionary, stop
    if current_word not in markov_chain:
        break

    # Get a random word from the list of next words
    next_word = random.choice(markov_chain[current_word])

    # Add the word to the generated text
    generated_text.append(next_word)

    # Set the next word as the current word for the next iteration
    current_word = next_word

# Join the list of words into a single string
generated_text = ' '.join(generated_text)
print(generated_text)
```
